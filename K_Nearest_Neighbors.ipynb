```python
import tensorflow
import keras
import sklearn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle

from sklearn import preprocessing
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
```

```python
data = pd.read_csv("./data/heart_failure_clinical_records_dataset.csv")
data.head()
```

```python
le = preprocessing.LabelEncoder()

age = le.fit_transform(list(data["age"]))
anaemia = le.fit_transform(list(data["anaemia"]))
creatinine_phosphokinase = le.fit_transform(list(data["creatinine_phosphokinase"]))
diabetes = le.fit_transform(list(data["diabetes"]))
ejection_fraction = le.fit_transform(list(data["ejection_fraction"]))
high_blood_pressure = le.fit_transform(list(data["high_blood_pressure"]))
platelets = le.fit_transform(list(data["platelets"]))
serum_creatinine = le.fit_transform(list(data["serum_creatinine"]))
serum_sodium = le.fit_transform(list(data["serum_sodium"]))
sex = le.fit_transform(list(data["sex"]))
smoking = le.fit_transform(list(data["smoking"]))
time = le.fit_transform(list(data["time"]))
DEATH_EVENT = le.fit_transform(list(data["DEATH_EVENT"]))

predict = "DEATH_EVENT"

X = list(zip(age, anaemia, creatinine_phosphokinase, diabetes, ejection_fraction, high_blood_pressure, platelets, serum_creatinine, serum_sodium, sex, smoking, time))
Y = list(DEATH_EVENT)

x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X, Y, test_size=0.3)

model = KNeighborsClassifier(n_neighbors=5)
model.fit(x_train, y_train)

acc = model.score(x_test, y_test)
print(acc)
```

```python
best = acc

for i in range(1000):
    x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X, Y, test_size=0.3)

    model = KNeighborsClassifier(n_neighbors=5)
    model.fit(x_train, y_train)

    acc = model.score(x_test, y_test)
    
    if acc > 0.9:
        print(acc)
        best = acc
        with open("model/KNN_model.pickle", "wb") as f:
            pickle.dump(model, f)

print(best)
```
